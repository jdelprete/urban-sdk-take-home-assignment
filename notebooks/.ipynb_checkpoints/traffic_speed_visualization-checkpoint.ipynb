````python
import os

import pandas as pd
import requests
from IPython.display import HTML, display
import IPython.core.display as core_display

# mapboxgl 0.10 expects display helpers on IPython.core.display.
core_display.HTML = HTML
core_display.display = display

from mapboxgl.utils import create_color_stops
from mapboxgl.viz import ChoroplethViz

BASE_URL = os.getenv("BASE_URL", "http://localhost:8000")
MAPBOX_TOKEN = os.getenv("MAPBOX_TOKEN")

if not MAPBOX_TOKEN:
    raise RuntimeError(
        "MAPBOX_TOKEN is not set. Export a valid Mapbox token before running this notebook."
    )

BASE_URL
````

````python
params = {"day": "Monday", "period": "AM Peak"}

response = requests.get(f"{BASE_URL}/aggregates/", params=params, timeout=60)
response.raise_for_status()
geojson_data = response.json()

len(geojson_data)
````

````python
features = [
    {
        "type": "Feature",
        "geometry": feature["geometry"],
        "properties": {
            "average_speed": feature["average_speed"],
            "freeflow_speed": feature["freeflow_speed"],
            "road_name": feature["road_name"],
            "link_id": feature["link_id"],
            "length_miles": feature["length_miles"],
        },
    }
    for feature in geojson_data
]

feature_collection = {"type": "FeatureCollection", "features": features}
feature_collection["features"][0]
````

````python
viz = ChoroplethViz(
    feature_collection,
    access_token=MAPBOX_TOKEN,
    color_property="average_speed",
    color_stops=create_color_stops([10, 20, 30, 40, 50], colors="Reds"),
    center=(-81.6557, 30.3322),
    zoom=11,
    line_width=2,
    opacity=0.85,
)
viz.show()
````

````python
df = pd.DataFrame(
    {
        "link_id": feature["link_id"],
        "road_name": feature["road_name"],
        "avg_speed": feature["average_speed"],
        "freeflow_speed": feature["freeflow_speed"],
        "length_miles": feature["length_miles"],
    }
    for feature in geojson_data
)

df.head()
````

````python
summary = pd.DataFrame(
    {
        "metric": [
            "segment_count",
            "mean_speed",
            "median_speed",
            "slowest_speed",
            "fastest_speed",
        ],
        "value": [
            len(df),
            round(df["avg_speed"].mean(), 2),
            round(df["avg_speed"].median(), 2),
            round(df["avg_speed"].min(), 2),
            round(df["avg_speed"].max(), 2),
        ],
    }
)

summary
````
